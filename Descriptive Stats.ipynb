```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df['SalePrice']
```

```python
median = np.median(df['SalePrice'])
median
```

```python
q1 = df['SalePrice'].quantile(0.25)
q3 = df['SalePrice'].quantile(0.75)

print(q1)
print(q3)
```

```python
q3-q1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(df['SalePrice'])
plt.show()
```

```python
outlier_lower_limit = q1 - 1.5*(q3-q1)
outlier_upper_limit = q3 + 1.5*(q3-q1)

print(outlier_lower_limit)
print(outlier_upper_limit)
```

```python
sales_price = df['SalePrice']

lower_limit_outlier = sales_price[sales_price < outlier_lower_limit].count()
upper_limit_outlier = sales_price[sales_price > outlier_upper_limit].count()

total = upper_limit_outlier+lower_limit_outlier
```
