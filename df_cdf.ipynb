```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample1 = np.random.normal(20, 1, 100000)
sample2 = np.random.normal(20, 3, 100000)
sample3 = np.random.normal(20, 10, 100000)

plt.hist(sample1, density=True, histtype='step', bins=100)
plt.hist(sample2, density=True, histtype='step', bins=100)
plt.hist(sample3, density=True, histtype='step', bins=100)

_ = plt.legend(('std = 1', 'std=3', 'std=10'))
plt.ylim(-0.01, 0.50)
plt.show()
```

```python
def cdf(dt):
    x = np.sort(dt)
    n = x.size
    y = np.arange(1, n+1)/n
    return (x,y)

x_sample1, y_sample1 = cdf(sample1)
x_sample2, y_sample2 = cdf(sample2)
x_sample3, y_sample3 = cdf(sample3)

plt.plot(x_sample1, y_sample1, marker='.', linestyle='None')
plt.plot(x_sample2, y_sample2, marker='.', linestyle='None')
plt.plot(x_sample3, y_sample3, marker='.', linestyle='None')
plt.plot()


```
